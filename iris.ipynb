```python
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import seaborn as sns
df = pd.read_csv("iris.csv")
```

```python
tmp = df.drop('Id', axis=1)
#we dont need id for plot
g = sns.pairplot(tmp, hue='Species', markers='*')
plt.show()
```

```python
#The above plot gives us the info that blue markers (Iris-Setosa) do not align/ dont have much in common with the other two species
```

```python
#Violin Plots for Fun, cause they look cool
Plot= sns.violinplot(y="Species",x="SepalLengthCm", data=df, inner="quartile")
plt.show()
```

```python
g = sns.violinplot(y='Species', x='SepalWidthCm', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='PetalLengthCm', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='PetalWidthCm', data=df, inner='quartile')
plt.show()
```
